```python
#import all modules and packeages needed
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#load in the dataset that I want to use
nutrition = pd.read_csv('Fast Foods - fastfoods.csv')

#show the dataset
display(nutrition)

#take two entire columns(will use their data in next line)
nutrition_protein = nutrition['Protein (g)'].to_numpy()
nutrition_satfats = nutrition['Saturated Fats (g)'].to_numpy()
nutrition_cholestral = nutrition['Cholesterol (mg)'].to_numpy()
nutrition_sodium = nutrition['Sodium (mg)'].to_numpy()
nutrition_calories = nutrition['Calories'].to_numpy()


#use data to put them in a scatter plot
plt.scatter(nutrition_protein, nutrition_satfats)
#change axis names
plt.xlabel('Protein (g)')
plt.ylabel('Saturated Fats (g)')
plt.show()

#use scatter plot to compate protein with cholestral
plt.plot(nutrition_protein, nutrition_cholestral, color = 'yellow')

#change axis names
plt.xlabel('Protein (g)')
plt.ylabel('Cholestral (mg)')

#seperate healthy & unhealthy contents in data with dashed line
plt.axhline(y = 159, color = 'r', linestyle = 'dashed')  
plt.show()
plt.bar(nutrition['Protein (g)'], nutrition['Calories'], align='center', color = 'red')

plt.xlabel('Protein (g)') 
plt.ylabel('Calories')
plt.show()

# draw lineplot
sns.lineplot(x="Protein (g)", y="Sodium (mg)", data=nutrition, color = 'green')

# specifying horizontal line type 
plt.axhline(y = 500, color = 'b', linestyle = 'dashed') 
plt.show()
plt.show()
```
